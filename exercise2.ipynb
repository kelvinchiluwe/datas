```python
import pandas as pd
import matplotlib.pyplot as plt

# Count occurrences of each country
country_counts = df['Country'].value_counts()

# Extract values for plotting
x = country_counts.index  # Country names
y = country_counts.values  # Frequency counts

# Create a single subplot
fig, ax = plt.subplots(figsize=(8, 6))

# Plot country counts
ax.plot(x, y, color='blue', marker='o', linestyle='--', markerfacecolor='g')
ax.set(title='Country business trend', xlabel='Country', ylabel='Count')

# Fix x-axis labels rotation
ax.set_xticks(range(len(x)))
ax.set_xticklabels(x, rotation=90)

# Enable grid for better visualization
ax.grid(True)

# Show plot
# plt.tight_layout()  # Prevents overlapping labels
plt.show()
```
